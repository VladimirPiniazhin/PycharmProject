```python
import matplotlib.pyplot as plt
from math import sin,cos,radians
```

```python
g = 9.81 #painovoiman kiihtyvyys m/s^2
dt = 0.05 #aikavälin pituus s
```

```python
x0 = 0.0 #Lähtöpisteen vaakakoordinaati
y0 = 1.8 #Lähtöjorkeus
v0 = 13 #Lähtönopeus
kulma = radians(35) #Lähtökulma (asteet sulkiujen sisään)
```

```python
vx = v0*cos(kulma)
vylist = [v0*sin(kulma)]
xlist = [x0]
ylist = [y0]
```

```python
while(ylist[-1] > 0):
    vylist.append(vylist[-1]-g*dt)
    xlist.append(xlist[-1]+vx*dt)
    vky = (vylist[-1]+vylist[-2])/2
    ylist.append(ylist[-1]+vky*dt)
```

```python
plt.plot(xlist,ylist)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.show()
```

```python
print("Lakikorkeus on", max(ylist), "m")
print("Kuulantyöntäjän tulos on", xlist[-1], "m")
```
