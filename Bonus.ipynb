```python
import matplotlib.pyplot as plt
```

```python
dt = 0.05 # aikavälin pituus (s)

time = [0.0] # ensimmäinen alkio (s)
acc = [3.0] # ensimmäinen alkio (m/s^2)
x = [0.0] # ensimmäinen alkio (m)

while (time[-1] + dt < 10.0):
    time.append(time[-1] + dt)
    if (time[-1] < 2.0):
        acc.append(3.0)
    elif (time[-1] < 7.0):
        acc.append(0.0)
    else:
        acc.append(-2.0)
```

```python
for i in range(5): # tulostetaan listojen 5 ensimmäistä alkiota
    print('Aika =',time[i],'s, Kiihtyvyys =',acc[i],'m/s^2')

print('....')

for i in range(len(time)-5,len(time)): # tulostetaan listojen 5 viimeistä alkiota
    print('Aika =',time[i],'s, Kiihtyvyys =',acc[i],'m/s^2')
```

```python
plt.xlabel('t (s)')
plt.ylabel('a (m/s$^{2}$)')
plt.plot(time,acc)
plt.show()
```

```python
vel = [0] # nopeuslistan ensimmäinen alkio (m/s)
for i in range(len(acc)-1):
    ak = (acc[i]+acc[i+1])/2 # aikavälin keskikiihtyvyys
    vel.append(vel[-1] + ak*dt)
```

```python
plt.xlabel('t (s)')
plt.ylabel('v (m/s)')
plt.plot(time,vel)
plt.show()
```
